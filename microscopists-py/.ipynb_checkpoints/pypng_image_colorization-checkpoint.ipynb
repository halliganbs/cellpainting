```python
def load_images(directory):
    # Loading filenames from directory
    list_files = os.listdir(directory)
    # remove the hidden file from the list
    list_files.pop(0)
    regex = re.compile(r'^(W\d{4}F\d{4}T\d{4}Z\d{3}C[1|3])\.tif$') # dont question it
    filtered = [i for i in list_files if regex.match(i)]
    dic = {}
    for c1, c3 in zip(filtered[0::2], filtered[1::2]):
        wf = c1[0:10]
        dic[wf] = (c1, c3)
    
    return dic


temp = load_images('images/projection')
```

```python
def make_images(dic):
    
    for wf in dic:
        nucleus = io.imread('images/projection/'+dic[wf][0], plugin="tifffile")
        body = io.imread('images/projection/'+dic[wf][1], plugin="tifffile")
        green = np.zeros(shape=np.shape(body)) # background
        
        out = np.zeros((body.shape[0], body.shape[1], 3), 'uint16')
        out[:,:,0] = nucleus # red
        out[:,:,1] = green
        out[:,:,2] = body # blue
        imageio.imwrite('output/'+wf+'.png', out)
        
```

```python
make_images(temp)
```
