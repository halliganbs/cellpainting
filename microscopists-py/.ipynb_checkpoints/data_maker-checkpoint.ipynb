```python
'''
Notebook to convert the images to .csv

halliganbs
'''

import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import png
import re #REEEEE

from skimage import io
```

```python
list_files = os.listdir('images/projection')
list_files
```

```python
def load_c1(directory, out_dir):
    list_files = os.listdir(directory)
    list_files.pop(0)
    regex = re.compile(r'^(W\d{4}F\d{4}T\d{4}Z\d{3}C[1])\.tif$')
    filtered = [i for i in list_files if regex.match(i)]
    dic = {}
    for c1 in zip(filtered[0::2]):
        wf = c1[0:10]
        dic[wf] = c1
    for wf in dic:
        print(dic[wf])
        filename = ''.join([directory, dic[wf][0]])
        im = io.imread(filename, plugin='tifffile')
        blank = np.zeros(shape=np.shape(im))
        
        colored = np.zeros((im.shape[0], im.shape[1], 3), 'uint16')
        colored[:,:,0] = im # red
        colored[:,:,1] = blank # green
        colored[:,:,2] = blank # blue
        io.imsave(out_dir+wf[0]+'.png', colored)

def load_c3(directory, out_dir):
    list_files = os.listdir(directory)
    list_files.pop(0)
    regex = re.compile(r'^(W\d{4}F\d{4}T\d{4}Z\d{3}C[3])\.tif$')
    filtered = [i for i in list_files if regex.match(i)]
    dic = {}
    for c3 in zip(filtered[0::2]):
        wf = c3[0:10]
        dic[wf] = c3
    
    for wf in dic:
        filename = ''.join([directory, dic[wf][0]])
        im = io.imread(filename, plugin='tifffile')
        blank = np.zeros(shape=np.shape(im))
        
        colored = np.zeros((im.shape[0], im.shape[1], 3), 'uint16')
        colored[:,:,0] = blank # red
        colored[:,:,1] = blank # green
        colored[:,:,2] = im # blue
        io.imsave(out_dir+wf[0]+'.png', colored)
```

```python
load_c1('images/projection/', 'temp/c1')
```

```python
load_c3('images/projection/', 'temp/c3/')
```
